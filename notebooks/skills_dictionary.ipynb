```python
import pandas as pd
from skills import SkillVocabularyBuilder, DEFAULT_SYNONYM_MAP

processed_dir = "../data/processed"
full_jobs = pd.read_csv(f"{processed_dir}/jobs_merged_cleaned.csv")

print(full_jobs.shape)
print(full_jobs.columns.tolist())

# Build the skill vocabulary from the job skills column
builder = SkillVocabularyBuilder(synonym_map=DEFAULT_SYNONYM_MAP)
builder.fit_from_dataframe(full_jobs, skills_col="job_skills")

print(builder.top_skills(20))
saved = builder.save_outputs(output_dir=processed_dir, min_frequency=10)
print(saved)
```
